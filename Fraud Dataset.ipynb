```python
from sklearn.impute import SimpleImputer
mv = SimpleImputer(missing_values = np.NaN, strategy = "constant", fill_value = 'Not Applicable')
x = data.iloc[:,5].values
x = x.reshape(-1,1)
data.iloc[:,5] = mv.fit_transform(x)
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x = data.iloc[:,5].values
data.iloc[:,5] = le.fit_transform(x)
x = x.reshape(-1,1)
data
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x = data.iloc[:,6].values
data.iloc[:,6] = le.fit_transform(x)
x = x.reshape(-1,1)
data
```

```python
from sklearn.model_selection import train_test_split

x = data[['job_id','telecommuting','has_company_logo','has_questions','employment_type','required_experience','required_education']]
y = data[['fraudulent']]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33)

# print(len(x_train))
# print(len(x_test))
# print(len(y_train))
# print(len(y_test))
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
x_train_mms = mms.fit_transform(x_train)
x_test_mms = mms.transform(x_test)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train_ss = ss.fit_transform(x_train)
x_test_ss = ss.transform(x_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
x_train_pca = pca.fit_transform(x_train_ss)
x_test_pca = pca.transform(x_train_ss)

x_train_pca = pca.fit_transform(x_test_ss)
x_test_pca = pca.transform(x_test_ss)
```

```python
l = []
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt 

knn = KNeighborsClassifier(n_neighbors = 6)
knn.fit(x_train_mms, y_train)

prediction = knn.predict(x_test_mms)

accuracy = accuracy_score(y_test, prediction)
print(accuracy)
l.append(accuracy)

target_names = ['fraud','not fraud']
print(classification_report(y_test, prediction.reshape(-1,1), target_names=target_names))

cm = confusion_matrix(y_test, prediction.reshape(-1,1), normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
fig, ax = plt.subplots(figsize=(8,6))
disp.plot(ax=ax, cmap='Blues')
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.svm import SVC
import matplotlib.pyplot as plt

svc = SVC()
svc.fit(x_train_mms, y_train)

prediction = svc.predict(x_test_mms)

accuracy = accuracy_score(prediction, y_test)
print(accuracy)
l.append(accuracy)

target_names = ['fraud','not fraud']
print(classification_report(y_test, prediction.reshape(-1,1), target_names=target_names))

cm = confusion_matrix(y_test, prediction.reshape(-1,1), normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
fig, ax = plt.subplots(figsize=(8,6))
disp.plot(ax=ax, cmap='Blues')
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.naive_bayes import GaussianNB
import matplotlib.pyplot as plt

nb = GaussianNB()
nb.fit(x_train_ss, y_train)

prediction = nb.predict(x_test_ss)

accuracy = accuracy_score(prediction, y_test)
print(accuracy)
l.append(accuracy)

target_names = ['fraud', 'not fraud']
print(classification_report(y_test, prediction.reshape(-1,1), target_names=target_names))

cm = confusion_matrix(y_test, prediction.reshape(-1,1), normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=target_names)
fig, ax = plt.subplots(figsize=(8,6))
disp.plot(ax=ax, cmap='Blues')
```
